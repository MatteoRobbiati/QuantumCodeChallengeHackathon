```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

#TODO: normalize all plots (if necessary)

df = pd.concat([pd.read_csv(r"presenza_15_010824-140824.csv"),
                pd.read_csv(r"presenza_15_150824-310824.csv"),
                pd.read_csv(r"presenza_15_010924-140924.csv"),
                pd.read_csv(r"presenza_15_150924_300924.csv"),
                pd.read_csv(r"presenza_15_011024-081024.csv"),
               ]
              )
df['datetime'] = pd.to_datetime(df['datetime'])
df_sorted = df.sort_values(by='datetime')
df0 =df_sorted[df_sorted['areaAnalisi'].str.contains('Intero Comune')]
df0.plot(x="datetime", y="value")
```

```python
fig=plt.figure()
labels = [f'Cagliari - 00{i}' for i in range(1,5)]

for i, label in enumerate(labels):
    df = df_sorted[df_sorted['areaAnalisi'].str.contains(label)]
    if i == 0:
        ax = df.plot(x="datetime", y="value", label=f"ACE0{i+1}")
    else:
        df.plot(x="datetime", y="value", label=f"ACE0{i+1}", ax=ax)
plt.show();
```

```python
fig=plt.figure()
labels = [f'Cagliari - 00{i}' for i in range(5,9)]

for i, label in enumerate(labels):
    df = df_sorted[df_sorted['areaAnalisi'].str.contains(label)]
    if i == 0:
        ax = df.plot(x="datetime", y="value", label=f"ACE0{i+5}")
    else:
        df.plot(x="datetime", y="value", label=f"ACE0{i+5}", ax=ax)
plt.show();
```
